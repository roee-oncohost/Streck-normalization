```python
dilutions = list(proteins_df.Dilution.value_counts().index)
```

```python
dilution_proteins_dict = {}
for dilution in dilutions:
    dilution_proteins = proteins_df[proteins_df['Dilution'] == dilution]['SeqId']
    dilution_proteins_dict[dilution] = dilution_proteins
```

```python
values_by_dilution = {}
for dilution, columns in dilution_proteins_dict.items():
    dilution_edta_df = edta_df[columns]
    dilution_streck_df = streck_df[columns]
    dilution_edta_values = dilution_edta_df.values.flatten()
    dilution_streck_values = dilution_streck_df.values.flatten()
    X = dilution_streck_values.reshape(-1, 1)
    
    values_by_dilution[dilution] = {
        'edta_values': dilution_edta_values,
        'streck_values': dilution_streck_values,
    }

    
```

```python
for dilution, readings in values_by_dilution.items():
    print(f"{dilution}: {len(readings['edta_values'])}")
```

```python
from sklearn.model_selection import train_test_split

for dilution, readings in values_by_dilution.items():


    streck_train, streck_test, edta_train, edta_test = train_test_split(readings['streck_values'], readings['edta_values'], test_size=0.2, random_state=42)
    values_by_dilution[dilution]['streck_train'] = streck_train
    values_by_dilution[dilution]['streck_test'] = streck_test   
    values_by_dilution[dilution]['edta_train'] = edta_train
    values_by_dilution[dilution]['edta_test'] = edta_test
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error
# import matplotlib.pyplot as plt

def evaluate_model(model, x_test, y_test):
    """Calculate and print evaluation metrics"""
    y_pred = model.predict(x_test)
    mse = mean_squared_error(y_test, y_pred)
    rmse = np.sqrt(mse)
    mae = mean_absolute_error(y_test, y_pred)
    r2 = r2_score(y_test, y_pred)
    
    print(f"\nTest Set Evaluation:")
    print(f"Mean Squared Error (MSE): {mse:.4f}")
    print(f"Root Mean Squared Error (RMSE): {rmse:.4f}")
    print(f"Mean Absolute Error (MAE): {mae:.4f}")
    print(f"R² Score: {r2:.4f}")
    
    return {'mse': mse, 'rmse': rmse, 'mae': mae, 'r2': r2}
```

```python
for dilution in ['20', '0.5', '0.005']:
    x_train = values_by_dilution[dilution]['streck_train'].reshape(-1, 1)
    y_train = values_by_dilution[dilution]['edta_train']    
    x_test = values_by_dilution[dilution]['streck_test'].reshape(-1, 1)
    y_test = values_by_dilution[dilution]['edta_test']

    model = LinearRegression()
    model.fit(x_train, y_train)
    print(f"For dilution {dilution}:\n\n")
    evaluate_model(model, x_test, y_test)
    print("********************************************************************************************")
```
